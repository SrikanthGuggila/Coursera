#### Handling Missing values

```python
for column in data.columns.tolist():
    if len(data[data[column] == '?'])>0:
        data[column].replace('?',np.nan,inplace=True)
```

```python
data.head()
```

```python
missing_data = data.isnull()
missing_data.head()
```

```python
import numpy as np
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
avg_normalized_losses = data['normalized-losses'].astype(float).mean(axis=0)
avg_normalized_losses
```

```python
data['normalized-losses'].replace(np.nan,avg_normalized_losses,inplace=True)
```

```python
data.head()
```

```python
avg_bore = data['bore'].astype(float).mean()
data['bore'].replace(np.nan,avg_bore,inplace=True)
```

```python
avg_stroke = data['stroke'].astype(float).mean()
data['stroke'].replace(np.nan,avg_stroke,inplace=True)

avg_horsepower = data['horsepower'].astype(float).mean()
data['horsepower'].replace(np.nan,avg_horsepower,inplace=True)

avg_peak_rpm = data['peak-rpm'].astype(float).mean()
data['peak-rpm'].replace(np.nan,avg_peak_rpm, inplace=True)

avg_price = data['price'].astype(float).mean()
data['price'].replace(np.nan,avg_price, inplace=True)
```

```python
data.head()
```

```python
missing_data = data.isnull()
for column in missing_data.columns.tolist():
    print(column)
    print(missing_data[column].value_counts())
```

```python
data['normalized-losses'] = data['normalized-losses'].astype('int')
data[['bore', 'stroke']] = data[['bore','stroke']].astype('float')
data['price'] = data['price'].astype('float')
data['peak-rpm'] = data['peak-rpm'].astype('float')
```

```python
data.dtypes
```

```python
data['city-L/100km'] = 235/data['city-mpg']
data['highway-L/100km'] = 235/data['highway-mpg']
```

```python
data.head()
```

```python
data['length'] = data['length']/data['length'].max()
data['width'] = data['width']/data['width'].max()
data['height'] = data['height']/data['height'].max()
```

```python
data['horsepower'] = data['horsepower'].astype('int', copy = True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.hist(data['horsepower'])

plt.xlabel('horsepower')
plt.ylabel('Count')
plt.title('horsepower bins')
```

```python
bins = np.linspace(min(data['horsepower']),max(data['horsepower']),4)
bins
```

```python
group_names = ['Low','Medium','High']
```

```python
data['horsepower-bin'] = pd.cut(data['horsepower'],bins,labels=group_names,include_lowest=True)
data[['horsepower','horsepower-bin']].head()
```

```python
data['horsepower-bin'].value_counts()
```

```python
dummy_variable_1 = pd.get_dummies(data['fuel-type'])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas' : 'fuel_type_gas', 'diesel' : 'fuel_type_diesel'},inplace=True)
dummy_variable_1.head()
```

```python
data = pd.concat([data, dummy_variable_1],axis=1)
data.drop('fuel-type', axis =1, inplace=True)
```

```python
dummy_variable_2 = pd.get_dummies(data['aspiration'])
dummy_variable_2.head()
```

```python
dummy_variable_2.rename(columns={'std' : 'aspiration_std', 'turbo': 'aspiration_trubo'},inplace=True)
dummy_variable_2.head()
```

```python
data = pd.concat([data, dummy_variable_2], axis=1)
data.drop('aspiration',axis=1,inplace=True)
```
