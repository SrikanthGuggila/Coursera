```python
grouped_test2 = data[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head()
```

```python
grouped_test2.get_group('fwd')['price']
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'],grouped_test2.get_group('fwd')['price'],grouped_test2.get_group('rwd')['price'])
print('ANOVA   F =',f_val,'P =',p_val)
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'],grouped_test2.get_group('fwd')['price'])
print('ANOVA   F =',f_val,'P =',p_val)
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('rwd')['price'],grouped_test2.get_group('fwd')['price'])
print('ANOVA   F =',f_val,'P =',p_val)
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'],grouped_test2.get_group('rwd')['price'])
print('ANOVA   F =',f_val,'P =',p_val)
```
