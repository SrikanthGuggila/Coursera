```python
def Distplot(RedFunc,BluFunc,RedName,BluName,Title):
    plt.figure(figsize = (8,6))
    
    ax1 = sns.distplot(RedFunc,hist=False,color='r',label=RedName)
    ax2 = sns.distplot(BluFunc,hist=False,color='b',label=BluName)
    
    plt.title(Title)
    plt.xlabel('Price (in dollers)')
    plt.ylabel('Proportion of cars')
    
    plt.show()
    plt.close()
```

```python
def pollyplot(xtrain,xtest,ytrain,ytest,lr,ploy_transform):
    plt.figure(figsize = (8,6))
    
    xmax = max([xtrain.values.max(),xtest.values.max()])
    xmin = min([xtrian.values.min(),xtest.values.min()])
    
    x = np.arange(xmin,xmax,0.1)
    
    plt.plot(xtrain,ytrain,color = 'ro', label = 'Trianing data')
    plt.plot(xtest,ytest,color = 'bo', label = 'Tesing data')
    plt.plot(x, lr.predict(poly_transform.fit_transform(x.reshape(-1,1))),label = 'Predicted Function')
    plt.ylim([-10000,60000])
    plt.ylabel('Price')
    plt.legend()
```
