```python
msk = np.random.rand(len(data_cdf)) < 0.8
train = data_cdf[msk]
test = data_cdf[~msk]
```

```python
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
```

```python
lr4 = LinearRegression()
lr4.fit(train_x,train_y)
print('Coeff:',lr4.coef_)
print('Intercept:',lr4.intercept_)
```

```python
plt.figure(figsize = (8,6))
plt.scatter(data_cdf['ENGINESIZE'],data_cdf['CO2EMISSIONS'],color='blue')
plt.plot(train_x,lr4.coef_[0]*train_x+lr4.intercept_[0],color = 'r')
plt.xlabel('Engine Size')
plt.ylabel('Emissions')
```
